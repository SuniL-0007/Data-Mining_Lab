```python
import pandas as pd
```

```python
def calculate_mean(data):
    if not data:
        return None
    total = 0
    for item in data:
        total += item
    return total / len(data)

def calculate_median(data):
    if not data:
        return None

    sorted_data = sorted(data)
    n = len(sorted_data)

    if n % 2 == 1:
        return sorted_data[n // 2]
    else:
        mid1 = sorted_data[n // 2 - 1]
        mid2 = sorted_data[n // 2]
        return (mid1 + mid2) / 2

def calculate_mode(data):
    if not data:
        return []

    frequency = {}
    for item in data:
        if item in frequency:
            frequency[item] += 1
        else:
            frequency[item] = 1

    if not frequency:
        return []

    max_frequency = 0
    for count in frequency.values():
        if count > max_frequency:
            max_frequency = count

    modes = []
    for item, count in frequency.items():
        if count == max_frequency:
            modes.append(item)

    return sorted(modes)

data_values = [1, 2, 2, 3, 4, 4, 4, 5, 6]

mean_val = calculate_mean(data_values)
median_val = calculate_median(data_values)
mode_val = calculate_mode(data_values)

print(f"Data values: {data_values}")
print(f"Mean: {mean_val}")
print(f"Median: {median_val}")
print(f"Mode: {mode_val}")

data_values_even = [10, 20, 30, 40]
mean_val_even = calculate_mean(data_values_even)
median_val_even = calculate_median(data_values_even)
mode_val_even = calculate_mode(data_values_even)
print(f"\nData values (even count): {data_values_even}")
print(f"Mean: {mean_val_even}")
print(f"Median: {median_val_even}")
print(f"Mode: {mode_val_even}")

data_values_no_mode = [1, 2, 3, 4, 5]
mode_val_no_mode = calculate_mode(data_values_no_mode)
print(f"\nData values (no distinct mode): {data_values_no_mode}")
print(f"Mode: {mode_val_no_mode}")

data_values_multi_mode = [1, 1, 2, 3, 3, 4]
mode_val_multi_mode = calculate_mode(data_values_multi_mode)
print(f"\nData values (multiple modes): {data_values_multi_mode}")
print(f"Mode: {mode_val_multi_mode}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Load data
data = pd.read_csv("/content/Data Mining.csv")

# Calculate standard deviation
print("Standard Deviation of Each Attribute:\n")
for col in data.columns:
    std = data[col].std(skipna=True)
    print(f"{col}: {std:.2f}")

# Plot bell curves
plt.figure(figsize=(10, 6))

for col in data.columns:
    values = data[col].dropna()
    mean = values.mean()
    std = values.std()

    x = np.linspace(values.min(), values.max(), 100)
    y = norm.pdf(x, mean, std)

    plt.plot(x, y, label=f'{col}')

plt.title("Bell Curves for Each Attribute")
plt.xlabel("Values")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()
```
