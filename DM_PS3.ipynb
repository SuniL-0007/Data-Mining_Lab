```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(42)

data = pd.DataFrame({
    'Cat1': np.random.choice(['A', 'B'], 100),
    'Cat2': np.random.choice(['X', 'Y', 'Z'], 100),
    'Num1': np.random.randint(10, 100, 100),
    'Num2': np.random.randint(20, 200, 100)
})
```

```python
contingency_table = pd.crosstab(data['Cat1'], data['Cat2'])
print("\nContingency Table:")
print(contingency_table)
```

```python
observed = contingency_table.values

row_sum = observed.sum(axis=1)
col_sum = observed.sum(axis=0)
total = observed.sum()

expected = np.outer(row_sum, col_sum) / total

chi_square = ((observed - expected) ** 2 / expected).sum()

print("\nChi-Square Value:", chi_square)
```

```python
correlation = np.corrcoef(data['Num1'], data['Num2'])[0, 1]
print("\nCorrelation Coefficient (Num1 & Num2):", correlation)
```

```python
data['Num1_MinMax'] = (data['Num1'] - data['Num1'].min()) / \
                      (data['Num1'].max() - data['Num1'].min())

data['Num2_MinMax'] = (data['Num2'] - data['Num2'].min()) / \
                      (data['Num2'].max() - data['Num2'].min())
```

```python
data['Num1_Zscore'] = (data['Num1'] - data['Num1'].mean()) / data['Num1'].std()
data['Num2_Zscore'] = (data['Num2'] - data['Num2'].mean()) / data['Num2'].std()

print("\nNormalized Data (First 5 Rows):")
print(data[['Num1_MinMax', 'Num2_MinMax', 'Num1_Zscore', 'Num2_Zscore']].head())
```
